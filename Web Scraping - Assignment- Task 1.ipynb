```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
# urlwiki variable assign for url address wikipedia
url = "https://en.wikipedia.org/wiki/Main_Page"
page = requests.get(url)
```

```python
#getting a response from url
page
```

```python
page.text
```

```python
#webscarping the wikipedia with beautiful soup
soup = BeautifulSoup(page.content)
```

```python
soup
```

```python
# getting headers from wikipedia web
header=[]
for headtext in soup.find_all("h2",class_="mp-h2"):
    header.append(headtext.text)
header
```
