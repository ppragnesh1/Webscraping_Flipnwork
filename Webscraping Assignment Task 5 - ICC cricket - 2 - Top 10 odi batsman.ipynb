```python
#importing necessary library for webscraping
import requests
from bs4 import BeautifulSoup
```

```python
#assigning website address to URL variable and get the url response with requests library and check if it give a valid response
url = ("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
req = requests.get(url)
req
```

```python
# assigned soup variable and get the content of page
soup = BeautifulSoup(req.content)
```

```python
soup
```

```python
# get the position value from the top 10 odi batsman and append to the pos list till 10 items
pos =[]
for p in soup.find_all("span",class_=["rankings-block__pos-number","rankings-table__pos-number"])[:10]:
    i = p.text.strip()
    pos.append(i)
```

```python
pos
```

```python
# get the name value from the top 10 odi batsman and append to the name list till 10 items
name = []
for n in soup.find_all("td",class_="table-body__cell rankings-table__name name")[:10]:
    n = n.text.strip()
    name.append(n)
```

```python
# get the team value from the top 10 odi batsman and append to the team list till 10 items
team = []
for t in soup.find_all("td",class_="table-body__cell nationality-logo rankings-table__team")[:10]:
    t = t.text.strip()
    team.append(t)
```

```python
# get the rating value from the top 10 odi batsman and append to the rating list till 10 items
rating = []
for r in soup.find_all("td",class_=["u-text-left","table-body__cell rating"])[0:10]:
    r =r.text.strip()
    rating.append(r)
```

```python
# gethering all list value to Dataframe with bining all list into dictionary to Dataframe
import pandas as pd
df = pd.DataFrame({"POS":pos,"Player":name,"Team":team,"Rating":rating})
```

```python
# checking if all list has same number of values
print(len(pos),len(name),len(team),len(rating))
```
