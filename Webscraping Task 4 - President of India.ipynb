```python
# importing relavant library for webscraping
import requests
from bs4 import BeautifulSoup
```

```python
#assign variable for the website address and checked if we get the response or not
url = "https://presidentofindia.nic.in/former-presidents.htm"
red = requests.get(url)
red
```

```python
#assign variable for Beautifulsoup to get the content of the website
soup = BeautifulSoup(red.content)
soup
```

```python
PresidentName = []
for item in soup.find_all("h3"):
    item = item.text
    item = item.strip()
    item = item.split("(")
    print(item)
    PresidentName.append(item[0])
    
```

```python
PresidentName
```

```python
Term =[]
for service in soup.find_all("div",class_="presidentListing"):
    #print(service.text)
    service = service.text
    service = service.split("\n")
    Term.append(service[2].replace("Term of Office: ",""))
    
```

```python
print(len(Term),len(PresidentName))
```
