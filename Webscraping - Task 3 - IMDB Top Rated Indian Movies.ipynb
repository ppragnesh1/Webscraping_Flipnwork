# import necessary library

```python
import requests
from bs4 import BeautifulSoup
```

```python
#set the variable with url for web address of top rated indian movies
url = "https://www.imdb.com/india/top-rated-indian-movies/"
req = requests.get(url)
req  # testing the url response
```

```python
#setting soup varibale
soup = BeautifulSoup(req.content)
soup
```

# gethering data for Movies name and movie year with 

```python
moviesname = []
moviesyear = []
rating = []
for item in soup.find_all("td",class_="titleColumn"):
    item = item.text
    item = item.split("\n")
    year = "".join(number for number in item[3] if number.isnumeric())
    moviesname.append(item[2].strip())
    moviesyear.append(year)
```

```python
print(moviesname)
print(moviesyear)
```

```python
print(len(moviesname),len(moviesyear))
```

# scarping the data for rating

```python
rating=[]
for r in soup("td",class_="ratingColumn imdbRating"):
    r = r.text
    r = r.split("\n")
    rating.append(float(r[1]))
```

```python
print(rating)
```

```python
print(len(moviesname),len(moviesyear),len(rating))
```

# saving data to Dataframe by importing pandas library

```python
import pandas as pd
topratedindmovies = pd.DataFrame({"Movies Name":moviesname,"Release Year":moviesyear,"IMDB Rating":rating})
print(topratedindmovies)
```
